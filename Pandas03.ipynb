```python
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Receita Total
df['Valor Venda'].sum()
```

```python
#Custo Total
df['custo'] = df['Custo Unitário'].mul(df['Quantidade'])
round(df['custo'].sum(), 2)
```

```python
df.head(1)
```

```python
#Lucro Total
df['lucro'] = df['Valor Venda'] - df['custo']
round(df['lucro'].sum(), 2)
```

```python
#Coluna com total de dias para enviar produto
df['TempoEnvio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(1)
```

# **Média de Tempo Envio por Marca**

```python
#Extraindo os dias
df['TempoEnvio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
#Tipo da Coluna TempoEnvio
df['TempoEnvio'].dtypes
```

```python
#Média do temo envio por Marca
df.groupby('Marca')['TempoEnvio'].mean()
```

### **Missing Values**

```python
#Verificando dados faltantes
df.isnull().sum()
```

### **Lucro por Ano e Marca**

```python
df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum().reset_index()
lucro_ano
```

```python
#Total Produtos Vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Gráfico Total Produtos Vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total Produtos Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
df.groupby(df['Data Venda'].dt.year)['lucro'].sum().plot.bar(title='Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
df.groupby(df['Data Venda'].dt.year)['lucro'].sum()
```

```python
#Selecionando apenas a venda de 2009
df2009 = df[df['Data Venda'].dt.year  == 2009]
```

```python
df2009.head()
```

```python
df2009.groupby(df2009['Data Venda'].dt.month)['lucro'].sum().plot(title='Lucro x Mês')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
df2009.groupby('Marca')['lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
df2009.groupby('Classe')['lucro'].sum().plot.bar(title='Lucro x Classe')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
